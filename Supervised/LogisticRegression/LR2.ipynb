```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedGroupKFold
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=15)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
X
```

```python
model = LogisticRegression()
valid_penalty_mapping = {
    'newton-cg': ['l2'],
    'lbfgs': ['l2'],
    'liblinear': ['l1', 'l2'],
    'sag': ['l2'],
    'saga': ['l1', 'l2', 'elasticnet']
}

c_values = [100, 10, 1, 0.1, 0.01]
solver_options = list(valid_penalty_mapping.keys())
```

```python
params = []
for solver in solver_options:
    for penalty in valid_penalty_mapping[solver]:
        params.append({'penalty': [penalty], 'C': c_values, 'solver': [solver]})
```

```python
# grid search cv
cv = StratifiedGroupKFold(n_splits=5)
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv = cv, n_jobs= -1)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
import numpy as np

mean_scores = grid.cv_results_['mean_test_score']
print("Mean test scores:", mean_scores)
print("Are all NaN?", np.isnan(mean_scores).all())
```
