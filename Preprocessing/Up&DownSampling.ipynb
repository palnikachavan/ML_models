USING SMOTE (Synthetic Minority Oversampling Technique)

```python
from sklearn.datasets import make_classification
```

```python
X, y  = make_classification(n_samples=1000, n_redundant=0, n_features=2,n_clusters_per_class=1, weights=[0.90], random_state=42)
```

```python
import pandas as pd
df2 = pd.DataFrame(X, columns=['f1','f2'])
df1 = pd.DataFrame(y, columns=['target'])
```

```python
df = pd.concat([df1, df2], axis = 1)
df.head()
```

```python
df.target.value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x=df['f1'], y=df['f2'], hue=df['target'])
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
# transform the dataset
oversample = SMOTE()
X,y = oversample.fit_resample(df[['f1','f2']], df['target'])
```

```python
df2 = pd.DataFrame(X, columns=['f1','f2'])
df1 = pd.DataFrame(y, columns=['target'])
new_df = pd.concat([df1, df2], axis = 1)
new_df.head()
```

```python
sns.scatterplot(x=new_df['f1'], y=new_df['f2'], hue=new_df['target'])
plt.show()
```
