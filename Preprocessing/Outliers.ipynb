5 number summary and box plot

```python
import numpy as np
marks = [45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
minm, q1, median, q3, maxm = np.quantile(marks, [0.0,0.25,0.50,0.75,1.0])
print(minm, q1, median, q3, maxm)
```

```python
iqr = q3-q1
low = q1 - 1.5*iqr
high = q3 + 1.5*iqr
print(low, high)
```

```python
import seaborn as sns
sns.boxplot(marks)
```
